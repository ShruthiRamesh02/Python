6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.



```python
import numpy as np

def calculate_median(data):
    """
    Calculate the median value of a dataset.

    Parameters:
        data (array-like): The dataset for which to calculate the median.

    Returns:
        float: The median value of the dataset.
    """
    median_value = np.median(data)
    return median_value

# Example usage
data_example = [10, 15, 20, 25, 30]
median_result = calculate_median(data_example)
print("Median of the dataset:", median_result)
#Explanations.
 #i have defined a function caluclated_median it takes array like data as input and caluclates median using np.median
 #the functions return the caluclated median value
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
conditions = [
    (df['total_bedrooms'] <= 10),
    (df['total_bedrooms'] > 10) & (df['total_bedrooms'] < 1000),
    (df['total_bedrooms'] >= 1000)
]
categories = ['small', 'medium', 'large']

df['total_bedroom_size'] = np.select(conditions, categories)

print(df.head())

#Explaination
# I defined conditions based on the given criteria 'total_bedrooms' into 'small', 'medium', or 'large' categories.
# I used numpy's select function to apply these conditions and assign the appropriate category to the 'total_bedroom_size' column.
```
