```python
class Human:   #Parent/base class
    def work(self):
        print("I can work")

class Male(Human):  #child/derived/sub class
    def dance(self):
        print("I can dance")
    def work(self):  #method overriding
        super().work()
        print("I can even code")

male1 = Male()
male1.dance()
male1.work()
```

```python
class Human:   #Parent/base class
    def __init__(self):
        self.num_eyes = 2
        self.num_nose = 1

    def work(self):
        print("I can work")

class Male(Human):  #child/derived/sub class
    def __init__(self):
        super().__init__()
        print("hello")
    def dance(self):
        print("I can dance")
    def work(self):  #method overriding
        super().work()
        print("I can even code")

male1 = Male()
male1.dance()
male1.work()
male1.num_eyes
```

```python
class Bank:
    def __init__(self):
        self.interestrate = 5
    def getinterestrate(self):
        print("The interest rate is: ", self.interestrate)

class PrivateBank(Bank):
    def __init__(self,loanamount):
        super().__init__()
        self.loanamount = loanamount
    def getminbalance(self):
        print("the min balance is 0")

class AxisBank(PrivateBank):
    def __init__(self,name,age):
        super().__init__(1500000)
        self.name = name
        self.age = age
        self.currentbalace = 20000
    def getCurrentbalance(self):
        print("The current balace is: ", self.currentbalace)
    def getMinbalance(self):
        print("The min balance is 5000")


bank = Bank()
bank.getinterestrate()

axisbank = AxisBank('Bindu',25)
axisbank.name
axisbank.getCurrentbalance()
axisbank.getMinbalance()
axisbank.getinterestrate()
axisbank.getminbalance()
```
