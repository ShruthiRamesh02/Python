12.Please convert BMI column into categories. For example, if BMI is < 19.5 code it as under weight, if weight is >= 19.5 and <= 25 code it as normal weight, if weight is >25 and <=30 code it as overweight and >30 code it as Obese.

````python
# Define function to categorize BMI
def categorize_bmi(bmi):
    if bmi < 19.5:
        return 'Underweight'
    elif 19.5 <= bmi <= 25:
        return 'Normal weight'
    elif 25 < bmi <= 30:
        return 'Overweight'
    else:
        return 'Obese'

# Apply the function to create a new column 'BMI categories'
df['BMI categories'] = df['Body mass index'].apply(categorize_bmi)
print(df.head())
````

13.Please determine which season had more absences in a given absenteeism dataset? Please tell us the name of season not number

````python
# Group by season and sum the absences
season_absences = df.groupby('Seasons')['Absenteeism time in hours'].sum()

# Find the season with the most absences
most_absent_season = season_absences.idxmax()

# Convert season number to season name
season_names = {1: 'Summer', 2: 'Autumn', 3: 'Winter', 4: 'Spring'}
most_absent_season_name = season_names[most_absent_season]

print(season_absences)
print(f'The season with the most absences was {most_absent_season_name}')
````

15.Find all words in a string that start with a vowel.
string = 'Apples are red, bananas are yellow, and oranges are orange.

````python
import re

string = 'Apples are red, bananas are yellow, and oranges are orange.'

# Find all words that start with a vowel
vowel_words = re.findall(r'\b[aeiouAEIOU]\w+', string)

print(vowel_words)
````

16.Replace all occurrences of the word "color" with "colour" in a string using regular expressions.
string = 'The color of the sky is blue.'

````python
import re

string = 'The color of the sky is blue.'

# Replace "color" with "colour" using regular expression
updated_string = re.sub(r'\bcolor\b', 'colour', string)

print(updated_string)
````

17.Extract all email addresses from a string and print them to the console.
string = 'Please contact us at support@example.com for assistance or info@example.com for more information.'

````python
import re

string = 'Please contact us at support@example.com for assistance or info@example.com for more information.'

# Regular expression pattern to match email addresses
pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

# Extract email addresses using regular expression
emails = re.findall(pattern, string)

for email in emails:
    print(email)
````

18.Validate a US phone number using regular expressions. The phone number should be in the format "XXX-XXX-XXXX" (where X is a digit).
phone_number = '123-456-7890'

````python
import re

phone_number = '123-456-7890'

# Regular expression pattern to match US phone numbers in the format XXX-XXX-XXXX
pattern = r'^\d{3}-\d{3}-\d{4}$'

if re.match(pattern, phone_number):
    print('Valid phone number')
else:
    print('Invalid phone number')
````

19.Parse a string of comma-separated key-value pairs and create a dictionary from them.
string = 'key1=value1,key2=value2,key3=value3'

````python
string = 'key1=value1,key2=value2,key3=value3'

# Split the string into key-value pairs using comma as the delimiter
pairs = string.split(',')

# Create a dictionary from the key-value pairs
dictionary = {}
for pair in pairs:
    key, value = pair.split('=')
    dictionary[key] = value

print(dictionary)
````

21.Calculate the time difference between two datetime objects in hours and minutes and print it to the console.
start_time = datetime(2023, 3, 24, 16, 30, 0), end_time = datetime(2023, 3, 24, 18, 45, 0)

````python
from datetime import datetime

start_time = datetime(2023, 3, 24, 16, 30, 0)
end_time = datetime(2023, 3, 24, 18, 45, 0)

# Calculate time difference
time_difference = end_time - start_time

# Calculate hours and minutes
hours = time_difference.seconds // 3600
minutes = (time_difference.seconds // 60) % 60

print(f'The time difference between {start_time} and {end_time} is {hours} hours and {minutes} minutes.')
````
