```python
import math

class Shape:
    def __init__(self, name, color):
        self.name = name
        self.color = color

    def area(self):
        pass

    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius, color):
        super().__init__('circle', color)
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius**2, 2)

    def perimeter(self):
        return round(2 * math.pi * self.radius, 2)

class Square(Shape):
    def __init__(self, side_length, color):
        super().__init__('square', color)
        self.side_length = side_length

    def area(self):
        return self.side_length**2

    def perimeter(self):
        return 4 * self.side_length

class Rectangle(Shape):
    def __init__(self, length, width, color):
        super().__init__('rectangle', color)
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create instances of each shape
circle = Circle(radius=5, color='red')
square = Square(side_length=6, color='blue')
rectangle = Rectangle(length=8, width=4, color='green')

# Print out their area and perimeter
print(f"{circle.name} area is {circle.area()}, perimeter is {circle.perimeter()}")
print(f"{square.name} area is {square.area()}, perimeter is {square.perimeter()}")
print(f"{rectangle.name} area is {rectangle.area()}, perimeter is {rectangle.perimeter()}")
```
