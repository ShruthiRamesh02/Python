Data Split: 4:1 for training/validation

-- 1. Number of corona patients with shortness of breath
SELECT COUNT(*) FROM corona_tested_006 WHERE `Shortness of breath` = 'true';

-- 2. Number of negative corona patients with fever and sore throat
SELECT COUNT(*) FROM corona_tested_006 WHERE `Fever` = 'true' AND `Sore throat` = 'true' AND `Covid report` = 'Negative';

-- 3. Rank the number of positive cases by month
SELECT MONTH(`Test date`) AS TestMonth, COUNT(*) AS PositiveCases
FROM corona_tested_006
WHERE `Covid report` = 'Positive'
GROUP BY TestMonth
ORDER BY PositiveCases DESC;

-- 4. Female negative corona patients with cough and headache
SELECT COUNT(*) FROM corona_tested_006 WHERE `Sex` = 'female' AND `Cough` = 'true' AND `Headache` = 'true' AND `Covid report` = 'Negative';

-- 5. Elderly corona patients with breathing problems
SELECT COUNT(*) FROM corona_tested_006 WHERE `Age ≥60 above years` = 'true' AND `Shortness of breath` = 'true';

-- 6. Three most common symptoms among COVID positive patients
SELECT `Cough`, `Fever`, `Sore throat`, COUNT(*) AS SymptomCount
FROM corona_tested_006
WHERE `Covid report` = 'Positive'
GROUP BY `Cough`, `Fever`, `Sore throat`
ORDER BY SymptomCount DESC
LIMIT 3;

-- 7. Less common symptom among COVID negative people
SELECT `Headache`, COUNT(*) AS SymptomCount
FROM corona_tested_006
WHERE `Covid report` = 'Negative' AND `Headache` = 'true';

-- 8. Most common symptoms among COVID positive males with known contact abroad
SELECT `Cough`, `Fever`, `Sore throat`, COUNT(*) AS SymptomCount
FROM corona_tested_006
WHERE `Covid report` = 'Positive' AND `Sex` = 'male' AND `Known contact with an individual confirmed to have COVID-19` = 'true'
GROUP BY `Cough`, `Fever`, `Sore throat`
ORDER BY SymptomCount DESC;
